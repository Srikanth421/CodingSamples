# Modelling

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
X = final_df.loc[:,'radius_mean':'fractal_dimension_worst'].values
y = final_df.loc[:,'diagnosis'].values
y = LabelEncoder().fit_transform(y)
```

```python
# Split data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,stratify=y,random_state=1)
```

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=100.0,random_state=1,solver='lbfgs',multi_class='ovr')
```

```python
lr.fit(X_train,y_train)
```

```python
# Check model built
lr.coef_
```

```python
from sklearn.feature_selection import RFE
rfe = RFE(lr,5)
fit = rfe.fit(X_train,y_train)
```

```python
# Number of features
fit.n_features_
```

```python
# Selected features
fit.support_
```

```python
# Feature ranking
fit.ranking_
```

```python
final_df.columns # top 5 features: texture_worst, perimeter_mean, smoothness_sd_error,fractal_dimension_mean,fractal_dimension_sd_error
```

```python
y_pred = lr.predict_proba(X_test)
```

```python
print('Test Accuracy: %.3f' % lr.score(X_test,y_test))
```

# If using logistic regression, we will use Log as the loss function since it has three advantages â€“ it gives probabilities which are easy to interpret, itâ€™s differentiable and lastly if a point is mis-classified the log(1 + exp(- h(x)y)) essentially is similar to log(exp(+h(x)y) or just h(x)y â€“ that is it becomes linear which is handy if there is some noise in the data. When it comes to regularization we would prefer l2 as l1 might remove some features and our dataset doesnâ€™t have many features. Another advantage of l2 s that itâ€™s strictly convex i.e., if a minimum exists it will be unique.

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(criterion='gini',max_depth=4,random_state=1)
```

```python
tree_model = tree_model.fit(X_train,y_train)
```

```python
y_pred = tree_model.predict(X_test)
```

```python
# Check Accuracy
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

# Explanation(Technical) - We built a logistic regression model with a pretty good accuracy but as mentioned above, using l2 reguralization might be helpful to check for overfitting.
